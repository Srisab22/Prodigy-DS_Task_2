```python
# Check for missing or null values in the dataset
df_titanic.isnull().sum().T
```

```python
#Fill Missing Values in the Age Column using median
df_titanic.fillna({"Age": df_titanic["Age"].median()}, inplace = True)
```

```python
#Drop Unnecessary Columns
df_titanic1 = df_titanic.drop(['PassengerId', 'Name', 'SibSp', 'Ticket', 'Cabin'], axis=1,)
df_titanic1.head(10)
```

```python
#Fill Missing Values in the Embarked Column using mode
df_titanic1.fillna({"Embarked": df_titanic1["Embarked"].mode()[0]}, inplace = True)
df_titanic1
```

```python
df_titanic1.tail(10)
```

```python

df_titanic1.columns
```

```python
#Check for Missing Values After Filling
df_titanic1.isnull().sum().T
```

```python
#Map Embarked Values to Numerical Categories
Embarked_mapping = {'S': 0, 'Q': 1, 'C': 2}
df_titanic1['Embarked'] = df_titanic1['Embarked'].map(Embarked_mapping)
df_titanic1
```

```python
#Convert Gender to Numerical Categories: 0 for male and 1 for female
df_new_titanic = df_titanic1["Sex"].apply(lambda x: 0 if x == "male" else 1)

df_new_titanic.head(10)
```

```python

df_titanic1["Sex"] = df_new_titanic
df_titanic1
```

```python
#calculating the correaltion of the given dataset
df_titanic1.corr()
```

```python
#plotting Correlation Heatmap
plt.figure(figsize=(13, 7))
sns.heatmap(df_titanic1.corr(), annot=True, fmt=".2f", cmap="inferno")
plt.title("Heatmap of Titanic dataset showing Correlation")
plt.show()
```

```python
#potting countplot of Survival Rate by Passenger Class
fig, ax = plt.subplots(figsize=(7, 6))
sns.countplot(data=df_titanic1, x='Pclass', hue='Survived', palette='magma', ax=ax)
ax.set_title('Survival Rate by Passenger Class')
plt.show()
```

```python
# plotting countplot of Survival Rate by Gender
fig, ax = plt.subplots(figsize=(7, 6))
sns.countplot(x="Sex", hue="Survived", data=df_titanic1, palette="coolwarm", ax=ax)
ax.set_title("Survival Rate by Gender")
plt.show()
```

```python
#plotting countplot of Survival Rate by Embarked Location
fig, ax = plt.subplots(figsize=(7, 6))
sns.countplot(data=df_titanic1, x="Embarked", hue="Survived", palette="viridis", ax=ax)
ax.set_title("Survival Rate by Embarked Location")
plt.show()
```

```python
#plotting kdeplot of Age Distribution by Survival
fig, ax = plt.subplots(figsize=(9, 6))
sns.kdeplot(data=df_titanic1[df_titanic1['Survived'] == 1], x='Age', fill=True, label='Survived', ax=ax)
sns.kdeplot(data=df_titanic1[df_titanic1['Survived'] == 0], x='Age', fill=True, label='Not Survived', ax=ax)
ax.set_title('Age Distribution by Survival')
ax.legend()
plt.show()
```

```python
#plotting catplot of Survival Rate by Passenger Class and Gender
sns.catplot(x='Pclass', hue='Survived', col='Sex', kind='count', data=df_titanic1, palette='magma', height=5, aspect=1)
fig.suptitle('Survival Rate by Passenger Class and Gender', y=1.03)
plt.show()
```

```python
#plotting Scatterplot of Age vs Fare by Survival
fig, ax = plt.subplots(figsize=(12, 7))
sns.scatterplot(data=df_titanic1, x='Age', y='Fare', hue='Survived', style='Survived', palette='Set1', ax=ax)
ax.set_title('Age vs Fare by Survival')
plt.show()
```

```python
#plotting Pairplot of All Variables
plt.figure(figsize=(20, 10))
sns.pairplot(data = df_titanic1, hue='Survived', palette="coolwarm")
plt.suptitle('Pairplot of All Variables', y = 1.02)
plt.show() 
```
